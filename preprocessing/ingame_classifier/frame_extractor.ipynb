```
import sys
import argparse

import cv2
print(cv2.__version__)
```

```
def extractImages(pathIn, pathOut, frame=60):
    vidcap = cv2.VideoCapture(pathIn)
    success,image = vidcap.read()
    count = 0
    success = True
    while success:
      success,image = vidcap.read()
      if(not success):
        break
      if(count % frame == 0):
        cv2.imwrite( pathOut + "frame%d.jpg" % count, image)     # save frame as JPEG file
      count += 1
    print("Extract done")
```
